# QUESTION THREE:Searching and Sorting Algorithms
Given the list 
arr = [21, 10, 45, 32, 67, 5, 18]

```python
arr = [21, 10, 45, 32, 67, 5, 18]
# a)Write a Python function to implement linear search to find the value 32
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

linear_search(arr, 32)
```

```python
# b)Write a Python function to implement binary search to search to find the value 45
arr_sorted = sorted(arr)

def binary_search(arr, target):
    low = 0
    high = len(arr) -1

    while low<= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid 
        elif arr[mid] < target:
            low = mid + 1
        else: 
            high = mid - 1
    return -1

binary_search(arr_sorted, 45)
```

```python
# c) Implement bubble sort to sort the list in ascending order
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
         for j in range(0, n-i-1):
             if arr[j] > arr[j+1]:
                 arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

bubble_sort(arr)
```
