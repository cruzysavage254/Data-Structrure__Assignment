Activity 2: Dictionary-Based Summary

```python
student_scores = {
    "Alice": [70, 80, 90],
    "Brian": [60, 75, 85],
    "Cynthia": [88, 92, 95]
}

summary = {}

for name, scores in student_scores.items():
    summary[name] = {
        "total": sum(scores),
        "average": sum(scores) / len(scores),
        "maximum": max(scores),
        }

print(summary),

for name, data in summary.items():
    print(f"\n Student: {name}")
    print(f" Total: {data['total']}")
    print(f" Average: {data['average']}")
    print(f" Maximum: {data['maximum']}")
```

Activity 3: Set Operations

```python
class_a = {"Alice", "Brian", "Cynthia"}
class_b = {"Brian", "David", "Emma"}

print("Results")

both_classes = class_a.intersection(class_b)
print("Students in both classes:", both_classes)

only_class_a = class_a.difference(class_b)
print("Students only in class A:", only_class_a)

all_students = class_a.union(class_b)
print("All students:", all_students)

```
